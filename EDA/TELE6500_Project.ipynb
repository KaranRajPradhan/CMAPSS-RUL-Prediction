```python


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


train_file = 'data/train_FD001.txt'
test_file = 'data/test_FD001.txt'
rul_file = 'data/RUL_FD001.txt'

column_names = [
    "EngineID", "Cycle", "Op1", "Op2", "Op3",
    "Sensor1", "Sensor2", "Sensor3", "Sensor4", "Sensor5",
    "Sensor6", "Sensor7", "Sensor8", "Sensor9", "Sensor10",
    "Sensor11", "Sensor12", "Sensor13", "Sensor14", "Sensor15",
    "Sensor16", "Sensor17", "Sensor18", "Sensor19", "Sensor20",
    "Sensor21"
]

# Load the training data
train_df1 = pd.read_csv("data/train_FD001.txt", sep='\s+', header=None, names=column_names)
train_df2 = pd.read_csv("data/train_FD002.txt", sep='\s+', header=None, names=column_names)
train_df3 = pd.read_csv("data/train_FD003.txt", sep='\s+', header=None, names=column_names)
train_df4 = pd.read_csv("data/train_FD004.txt", sep='\s+', header=None, names=column_names)

train_df1.head()
```

```python
# Find the rows corresponding to the maximum cycles for each engine
failure_data = train_df1.loc[train_df1.groupby('EngineID')['Cycle'].idxmax()]

# Plot the maximum cycles for each engine
plt.figure(figsize=(14, 10))
failure_data.set_index('EngineID')['Cycle'].plot(kind='bar')
plt.title('Cycles to Failure for Each Engine')
plt.xlabel('Engine Number')
plt.ylabel('Cycles to Failure')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Scatter plots for Op1, Op2, Op3 against largest cycle (failure point)
plt.figure(figsize=(14, 10))

plt.subplot(3, 1, 1)
sns.scatterplot(data=failure_data, x='Op1', y='Cycle', color='blue')
plt.title('Effect of Operational Setting 1 (Op1) on Failure Cycles')
plt.xlabel('Op1')
plt.ylabel('Cycles to Failure')

plt.subplot(3, 1, 2)
sns.scatterplot(data=failure_data, x='Op2', y='Cycle', color='green')
plt.title('Effect of Operational Setting 2 (Op2) on Failure Cycles')
plt.xlabel('Op2')
plt.ylabel('Cycles to Failure')

plt.subplot(3, 1, 3)
sns.scatterplot(data=failure_data, x='Op3', y='Cycle', color='red')
plt.title('Effect of Operational Setting 3 (Op3) on Failure Cycles')
plt.xlabel('Op3')
plt.ylabel('Cycles to Failure')

plt.tight_layout()
plt.show()
```

```python
# Distribution of Operational Settings
plt.figure(figsize=(14, 10))
for i, op in enumerate(["Op1", "Op2", "Op3"], start=1):
    plt.subplot(3, 1, i)
    sns.histplot(train_df1[op], kde=True, color=f"C{i}")
    plt.title(f"Distribution of {op}")
    plt.xlabel(op)
    plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(16, 12))
correlation_matrix = train_df1.drop(columns=["EngineID", "Cycle"]).corr()

sns.heatmap(
    correlation_matrix, 
    annot=True,
    fmt=".2f",
    cmap="coolwarm"
)
plt.title("Feature Correlation", fontsize=18)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```

```python
# Outlier Detection (Z-Score)
from scipy.stats import zscore

sensor_cols = train_df1.columns[5:]
train_df1["Outlier_Score"] = train_df1[sensor_cols].apply(zscore).mean(axis=1)
outliers = train_df1[train_df1["Outlier_Score"] > 3]

plt.figure(figsize=(12, 6))
sns.scatterplot(data=train_df1, x="Cycle", y="Outlier_Score", hue=(train_df1["Outlier_Score"] > 2))
plt.title("Outlier Detection (Z-Score)")
plt.xlabel("Cycle")
plt.ylabel("Outlier Score")
plt.legend(title="Is Outlier?")
plt.show()
```
